```python
states_of_interest = ['Alabama', 'California', 'Texas', 'Mississippi']
filtered_data = df[df['State'].isin(states_of_interest)]

years = list(range(2020, 1994, -1))
```

```python
plt.figure(figsize=(10, 6))

for state in states_of_interest:
    state_data = filtered_data[filtered_data['State'] == state]
    poverty_levels = state_data[state_data['Year'].isin(years)]['Percent in poverty']
    plt.plot(years, poverty_levels, label=state)

plt.title('Poverty Levels Over Years')
plt.xlabel('Year')
plt.ylabel('Poverty Level')
plt.legend()
plt.grid(True)
plt.savefig('Poverty4States.png')  # Change the file extension to save in a different format
plt.show()
```

```python
# Filter the DataFrame for the state of Texas
data_sub = df[df['State'] == 'Texas']['Percent in poverty']

# Perform seasonal decomposition for Texas
decomposition = sm.tsa.seasonal_decompose(data_sub, model='additive', period=12)

# Access the decomposed components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Define the Dickey-Fuller stationarity test function
def dickey_fuller(timeseries):
    print('Dickey-Fuller Stationarity test:')
    test = sm.tsa.adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic', 'p-value', 'Number of Lags Used', 'Number of Observations Used'])
    for key, value in test[4].items():
        result['Critical Value (%s)' % key] = value
    print(result)

# Apply the Dickey-Fuller test using the function on the time series
dickey_fuller(data_sub)

# Plot the separate components
decomposition.plot()
plt.grid(True)
plt.savefig('Decomp_Texas.png', transparent=False)
plt.show()
```
